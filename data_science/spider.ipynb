```python
import requests           #collects the webpage data object
from bs4 import BeautifulSoup  #convert the requests object in readable object
```

```python
addr='https://en.wikipedia.org/wiki/September_10/'  #page address
page=requests.get(addr)#collect the data from webpage
print(page.status_code)#status code 
```

```python
soup=BeautifulSoup(page.text) #soup->object of class BeautifulSoup
```

```python
soup
```

```python
soup.find('div',attrs={'id':'mp-otd'}).text
```

```python
datalist=soup.find_all('ul')
for item in datalist:
    print(item.text)
```
